Preparation Data

```python
import cv2 as cv
import matplotlib.pyplot as plt


image = cv.imread('/content/KTMNew.jpg')
image = cv.cvtColor(image, cv.COLOR_BGR2RGB)
resize_image = cv.resize(image, (0,0), fx=0.5, fy=0.5)
plt.imshow(resize_image)
plt.show()
```

Tutup bagian teks Politeknik Negeri Malang

```python
cv.rectangle(resize_image, pt1=(300,240), pt2=(670,400), color=(0,0,255), thickness=-1)
plt.imshow(resize_image)
plt.show()
```

Tutup bagian BNI dan Prodi

```python
cv.rectangle(resize_image,pt1=(1658,220),pt2=(1900,400),color=(0,255,255),thickness=-1)
cv.rectangle(resize_image,pt1=(510,1350),pt2=(1300,1400),color=(0,0,0),thickness=-1)
plt.imshow(resize_image)
plt.show()
```
